```python
from pyspark.sql import SparkSession
spark= SparkSession.builder.appName("DataEngineering").getOrCreate()
```

```python
# Reading the CUSTOMERS.csv file
# The 'header' option is set to true to use the first row as column names.
# The 'inferSchema' option is set to true to automatically detect the data types of each column.
df_customers=spark.read.format("csv").option("header", "true").option("inferSchema", "true").load("C:/Users/sukan/OneDrive/Documents/Masters/Projects/CUSTOMERS.csv")
# removing the dupliacte records
df_customers=df_customers.dropduplicates()
display(df_customers.limit(100))
```

```python
# Reading the CARS.csv file
# we use the header and inferSchema options for schema detection.
df_cars=spark.read.format("csv").option("header", "true").option("inferSchema", "true").load("C:/Users/sukan/OneDrive/Documents/Masters/Projects/CARS.csv")
df_cars=df_cars.dropDuplicates()
display(df_cars.limit(100))
```

```python
# Reading the HOUSEHOLDS.csv file
# This cell also uses header and inferSchema options to read the data with proper schema.
df_households=spark.read.format("csv").option("header", "true").option("inferSchema", "true").load("C:/Users/sukan/OneDrive/Documents/Masters/Projects/HOUSEHOLDS.csv")
df_households=df_households.dropDuplicates()
display(df_households.limit(100))
```

```python
# Joining all the 3 csv files 
joined_df=df_customers.join(df_households, df_customers.CUST_ID==df_households.CUST_ID)
joined_df=joined_df.join(df_cars, df_cars.CAR_ID==joined_df.CAR_ID)
joined_df.dispaly()
```

```python
# priniting schema of the dataframe
joined_df.printSchema()
```

```python
# priniting count of records
display(joined_df.count())
```

```python
# saving dataframe as parquet file
joined_df.write.parquet("/path/to/save/joined_df.parquet")
```
