```python
#combining the data.
combined_df = pd.merge(df23, df22, on='CUST_ID')
combined_df = pd.merge(combined_df, df21, on='CAR_ID')
combined_df.head()

```

```python
pd.set_option('display.max_columns', None)  # Show all columns
print(combined_df.head())
```

```python
#removing all the unnecessary columns above seen in the data frame
combined_df= combined_df.drop(columns=[col for col in combined_df.columns if "Unnamed" in col])
```

```python
#Data cleaning checklist
#1. check for Nan  Values
print("Checking for missing values in each column:")
print(combined_df.isnull().sum())  
```

```python
#Data cleaning checklist
#2. Check for Duplicate Rows:

print(f"Number of duplicate rows: {combined_df.duplicated().sum()}")  # Count duplicates
```

```python
#Data cleaning checklist
#3. Check for Data Type Mismatches
print("Data types before conversion:")
print(combined_df.dtypes)  # Display current data types

```

```python
#changing data types  of  HH Start Date, Date of Birth  from object to date 

combined_df['HH Start Date'] = pd.to_datetime(combined_df['HH Start Date'], errors='coerce')
combined_df['Date of Birth'] = pd.to_datetime(combined_df['Date of Birth'], errors='coerce')
print(combined_df.dtypes) #checking the modified data types
```
